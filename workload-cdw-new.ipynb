```python
import pandas as pd
import pyodbc
import random
import numpy as np
from datetime import timedelta
import sqlalchemy
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Total Compute Nodes-data-2024-04-25 03_42_44.csv')
df['Time'] = pd.to_datetime(df['Time'])
df
```

```python
# df.groupby([df['Time'].dt.date ,'totalComputeNodes'])['Time'].apply(lambda x: f'{x.min()} - {x.max() + pd.Timedelta(minutes=1)}')
```

```python
df = df.sort_values(by=['Time'])
df
```

```python
def roundNumber(x):
    if (x == 7 or x == 8 or x == 9)  :
        return 10
    elif (x == 11) :
        return 12
    elif (x == 13 or x == 15):
        return 16
    else:
        return x
        

df['totalComputeNodes'] = df['totalComputeNodes'].apply(roundNumber)
df
```

```python
checkWorkload = df.iloc[0 , 1]
checkDay = df.iloc[0 , 0].day
duration = []
durationMin = []
tmp = []
result = []
final = []
for i in range(len(df)):
    if(df.iloc[i , 1] == checkWorkload and df.iloc[i , 0].day ==  checkDay):
        if(i == len(df) - 1) :
            result.append(tmp)
        else:
            tmp.append([df.iloc[i , 0] , df.iloc[i , 1]])
    else:
        result.append(tmp)
        tmp = []
        tmp.append([df.iloc[i , 0] , df.iloc[i , 1]])
        checkWorkload = df.iloc[i , 1]
        checkDay = df.iloc[i , 0].day
```

```python
result
```

```python
for item in result:
    duration = f'{item[0][0].hour}:{item[0][0].minute} - {item[len(item) - 1][0].hour}:{item[len(item) - 1][0].minute + 4}'
    durationM = (((item[len(item) - 1][0] - item[0][0]).seconds) + 300) / 60
    for child in item :
        final.append(child + [duration, durationM])
    
```

```python
final
```

```python
final_pd = pd.DataFrame(final , columns = ['Time' , 'totalNodeRunning' , 'Duration' , 'DurationM'])
final_pd
```

```python
final_pd['VW'] = final_pd['totalNodeRunning'].apply(lambda x : 'User and Job' if int(x) > 10 else 'Job' )
final_pd
```

```python
def defineGroup(x):
    if (x == 6)  :
        return '1 group job'
    elif (x == 10) :
        return '2 group job'
    elif (x == 12):
        return '1 group job and 1 group user'
    elif (x == 16):
        return '2 group job and 1 group user'
    else :
        return '2 group job and 2 group user'
final_pd['group'] = final_pd['totalNodeRunning'].apply(defineGroup)
final_pd
```

```python
final_pd
```

```python
def changeTimeFormat(x):
    first = x.split('-')[0]
    first = first.split(":")
    first = [ f"{i}".strip() if int(i) > 9 else f'0{i}' for i in first]
    first = ':'.join(first)
    second = x.split('-')[1]
    second = second.split(":")
    second = [ f'{i2}'.strip() if int(i2) > 9 else f'0{i2.strip()}'.strip() for i2 in second]
    second = ':'.join(second)
    return f'{first} - {second}'

final_pd["Duration"] = final_pd["Duration"].apply(changeTimeFormat)
final_pd
```

```python
final_pd.to_csv('workload-cdp-ver2.csv' , index = False)
```
