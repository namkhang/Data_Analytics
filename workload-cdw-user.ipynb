```python
import pandas as pd
import random
import re
```

```python
df_raw_data = pd.read_csv("draw-file-workload-user.csv" , encoding = "ISO-8859-1" , header = None)
df_raw_data
```

```python
def ConvertColumnType(x):
    for i in range (len(x)) :
        x[i] = str(x[i])
    return x
df_raw_data = df_raw_data.apply(ConvertColumnType)
df_raw_data
```

```python
convertToList = list(df_raw_data.values)
convertToList = [' '.join(i).strip() for i in convertToList]
convertToList = [i.replace('nan' , '').strip() for i in convertToList]
convertToList
```

```python
df_raw_data = pd.DataFrame(convertToList , columns=['raw_data'])
df_raw_data
```

```python
df_raw_data = df_raw_data[(df_raw_data['raw_data'] != 'Missing Table Statistics') & (df_raw_data['raw_data'] != 'Slow Client')  & (df_raw_data['raw_data'] != 'Corrupt Table Statistics')  & (df_raw_data['raw_data'] != 'Slow Query Planning') & (df_raw_data['raw_data'] != 'Corrupt Table Statistics Slow Query Planning') & (df_raw_data['raw_data'] != 'Missing Table Statistics Slow Query Planning') & (df_raw_data['raw_data'] != 'Missing Table Statistics Slow Client') & (df_raw_data['raw_data'] != 'Slow Row Materialization') & (df_raw_data['raw_data'] != 'Slow Client Missing Table Statistics') & (df_raw_data['raw_data'] != 'Slow Client Slow Query Planning')]
df_raw_data
```

```python
data_source = []
tmp = []
count = 1
for i in range (len(df_raw_data)):
    if (count < 11 ):
        tmp.append(df_raw_data.iloc[i , 0])
        count+=1
    else:
        tmp.append(df_raw_data.iloc[i , 0])
        data_source.append(tmp)
        tmp = []
        count = 1
data_source
```

```python
df = pd.DataFrame(data_source , columns = ['Query' , 'Status' , 'Start Time' , 'Duration' , 'User' , 'Statement' , 'Rows Produced' , 'Input' , 'Output' , 'CPU' , 'Memory'])
df
```

```python
df = df.drop(columns = ['Statement' , 'Rows Produced' , 'Output' ])
df
```

```python
df = df.sort_values(by=['User' , 'Start Time' , 'Duration'])
df
```

```python
def getTimeUsed(x):
    used = 0
    x = x.split(" ")
    for it in x:
        if "h" in it:
            used += int(it.replace("h" , "")) * 60 * 60 * 1000
        elif "m" in it and "s" not in it :
            used += int(it.replace("m" , "")) * 60 * 1000
        elif "s" in it and "m" not in it:
            used += int(it.replace("s" , "")) * 1000
        else:
            used += int(it.replace("ms" , ""))
    return used / 60000
    
df['DurationM'] = df['Duration'].apply(getTimeUsed)
df
```

```python
df = df.sort_values(by = ['User' , 'Start Time' , 'DurationM'])
df
```

```python
df['Date'] = df['Start Time'].apply(lambda x : str(x).split(' ')[0])
df
```

```python
df_percent_usage = df.groupby(['User' , 'Date']).sum().reset_index().drop(["Query" , "Status" , "Start Time", "Duration","CPU","Input" , "Memory"] , axis = 1)
df_percent_usage
```

```python
arr_sum_by_day = []
checked = []
conver_pd_to_list = list(df_percent_usage.values)
for i in conver_pd_to_list:
    if i[1] not in checked:
        sum_tt = sum([it[2] for it in conver_pd_to_list if it[1] == i[1]])
        arr_sum_by_day.append([i[1] , sum_tt])
        checked.append(i[1])
    else:
        continue
arr_sum_by_day
```

```python
def countPercitent(x):
    sum_tt = [ix[1] for ix in arr_sum_by_day if ix[0] == x[1]][0]
    return (x[2] / sum_tt) * 100

df_percent_usage['percent'] = df_percent_usage.apply(countPercitent , axis = 1)
df_percent_usage
```

```python
df_percent_usage = df_percent_usage.sort_values(by = ['Date'])
df_percent_usage
```

```python
filter = []
conver_pd_to_list = list(df_percent_usage.values)
group_tmp = []
day = list(set(df['Date'].values))
for i in day  :
    list_day_match = [it for it in conver_pd_to_list if it[1] == i]
    for index in range(len(list_day_match)):
        if list_day_match[index][3] > 5:
            filter.append(list(list_day_match[index]))
        else :
            group_tmp.append(list(list_day_match[index]))
    if(len(group_tmp)) > 0 :
        filter.append(['Other' , group_tmp[0][1] ,sum([it1[2] for it1 in group_tmp]) , sum([it1[3] for it1 in group_tmp])])  
    group_tmp = []
filter
```

```python
df_percent_usage = pd.DataFrame(filter , columns = ['User' , 'Day', 'DurationM' , 'percent'])
df_percent_usage
```

```python
list_user = list((set(df['User'].values)))
user_format = []
for index in range(len(list_user)) :
    user_format.append({'name' : list_user[index] , 'alias' : index})
user_format
    
```

```python
def defineUser(x):
    return [ i['alias'] for i in user_format if x == i['name']][0]

df['alias'] = df['User'].apply(defineUser)
df
```
