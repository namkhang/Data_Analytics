```python
def convertColumnToArray(x):
    x = x.replace("[" , "").replace("]" , "")
    x = x.split(",")
    return x

real_time["frozen_times"] = real_time["frozen_times"].apply(convertColumnToArray)
real_time
```

```python
def seperateTime(x):
    if x[3][0] == "" :
        return [x[2]]
    else:
        parse_frozen_time = x[3].copy()
        parse_frozen_time.insert(0 , x[2])
        current_second = int(parse_frozen_time[0].split(" ")[1].split(":")[2].split(".")[0])
        current_milisecond = int(parse_frozen_time[0].split(" ")[1].split(".")[1])
        for i in range(len(parse_frozen_time)):
            if i == 0 :
                continue
            elif i == len(parse_frozen_time) - 1:
                if int(parse_frozen_time[i].split(" ")[1].split(":")[2].split(".")[0]) == current_second :
                    milisecond = int(int(current_milisecond) + 200)
                    parse_frozen_time[i] = f'{parse_frozen_time[i].split(".")[0]}.{milisecond}'
                    current_milisecond = int(parse_frozen_time[i].split(" ")[1].split(".")[1])
                else:
                    parse_frozen_time[i] = f'{parse_frozen_time[i].split(".")[0]}.000'
            else:
                if int(parse_frozen_time[i].split(" ")[1].split(":")[2].split(".")[0]) == current_second :
                    milisecond = int(int(current_milisecond) + 200)
                    parse_frozen_time[i] = f'{parse_frozen_time[i].split(".")[0]}.{milisecond}'
                    current_milisecond = int(parse_frozen_time[i].split(" ")[1].split(".")[1])
                else:
                        parse_frozen_time[i] = f'{parse_frozen_time[i].split(".")[0]}.000'
                        current_second = int(parse_frozen_time[i].split(" ")[1].split(":")[2].split(".")[0])
                        current_milisecond = 0
        return parse_frozen_time
real_time["parse_frozen_time"]=real_time.apply(seperateTime , axis = 1)
real_time

```
