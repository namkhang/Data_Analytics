```python
import pandas as pd
import random
import numpy as np
from datetime import timedelta
import math
```

```python
time = '20240431'
```

```python
sd1 = pd.read_csv(f"./800019/{time}/SD1.csv")
sd2 = pd.read_csv(f"./800019/{time}/SD2.csv")
sd3 = pd.read_csv(f"./800019/{time}/SD3.csv")
sd4 = pd.read_csv(f"./800019/{time}/SD4.csv")
sd7 = pd.read_csv(f"./800019/{time}/SD7.csv")
sd26 = pd.read_csv(f"./800019/{time}/SD26.csv")
```

```python
sd1['filename'] = sd1['filename'].apply(lambda x : '-'.join(x.split('-')[:2]))
sd2['filename'] = sd2['filename'].apply(lambda x : '-'.join(x.split('-')[:2]))
sd3['filename'] = sd3['filename'].apply(lambda x : '-'.join(x.split('-')[:2]))
sd4['filename'] = sd4['filename'].apply(lambda x : '-'.join(x.split('-')[:2]))
sd7['filename'] = sd7['filename'].apply(lambda x : '-'.join(x.split('-')[:2]))
sd26['filename'] = sd26['filename'].apply(lambda x : '-'.join(x.split('-')[:2]))
```

```python
sd1  = sd1.drop_duplicates()
sd2  = sd2.drop_duplicates()
sd3  = sd3.drop_duplicates()
sd4  = sd4.drop_duplicates()
sd7  = sd7.drop_duplicates()
sd26  = sd26.drop_duplicates()
```

```python
sd_merge = sd1.merge(sd2, how='left', on=['filename' , 'calendar_timestamp', 'YearHCode','YearLCode','MonthHCode','MonthLCode','DayHCode','DayLCode','HourHCode','HourLCode','MinuteHCode','MinuteLCode','SecondHCode','SecondLCode','MilisecondHCode','MilisecondLCode'])\
    .merge(sd3, how='left', on=['filename' , 'calendar_timestamp', 'YearHCode','YearLCode','MonthHCode','MonthLCode','DayHCode','DayLCode','HourHCode','HourLCode','MinuteHCode','MinuteLCode','SecondHCode','SecondLCode','MilisecondHCode','MilisecondLCode'])\
     .merge(sd4, how='left', on=['filename' , 'calendar_timestamp', 'YearHCode','YearLCode','MonthHCode','MonthLCode','DayHCode','DayLCode','HourHCode','HourLCode','MinuteHCode','MinuteLCode','SecondHCode','SecondLCode','MilisecondHCode','MilisecondLCode'])\
    .merge(sd7, how='left', on=['filename' , 'calendar_timestamp', 'YearHCode','YearLCode','MonthHCode','MonthLCode','DayHCode','DayLCode','HourHCode','HourLCode','MinuteHCode','MinuteLCode','SecondHCode','SecondLCode','MilisecondHCode','MilisecondLCode'])\
      .merge(sd26, how='left', on=['filename' , 'calendar_timestamp', 'YearHCode','YearLCode','MonthHCode','MonthLCode','DayHCode','DayLCode','HourHCode','HourLCode','MinuteHCode','MinuteLCode','SecondHCode','SecondLCode','MilisecondHCode','MilisecondLCode'])
sd_merge
```

```python
sd_merge.drop(['calendar_timestamp'] , axis = 1).to_csv(f'./result-800019/dev_data_800019_{time}.csv',index= False)
```
