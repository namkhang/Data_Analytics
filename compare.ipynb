```python
def decodeColumnByte(x):
    for i in range (len(x)) :
        if type(x[i]) == bytes :
            x[i] = x[i].decode("utf-8")
    return x
pre = pre.apply(decodeColumnByte)
pre
#because get data from parquet file so these column has been covert from string to bytes
# the fuction help decode it
```

```python
def convertStringtoInt(x):  
    for i in range (len(x)) :
        if type(x[i]) == str : 
            x[i] = x[i].replace('?' ,'').replace('#' ,'').replace('\x00\x00\x00\x00\x00\x00' , '').replace('\x00\x00\x00\x00' , '').replace('\x00\x00' , '')
            try:
                x[i] = int(x[i])
            except: 
                continue
    return x
pre = pre.apply(convertStringtoInt)
pre
#some null column in impala will be exported as '?' , '#' , '\x00\x00\' ,... so we need to replace it to blank
#when we export data, some int column  will be exported to string (example original data is 3456 and after exporting it will be '3456??')
# therefore we need to convert it back to original data type on impala (int) before comparing 
```

```python
pro = pro.apply(convertStringtoInt)
pro
```
